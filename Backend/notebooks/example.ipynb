```python
#AGE RANGE
df["BIRTHDATE"] = pd.to_datetime(df["BIRTHDATE"], errors="coerce")
df["AGE"] = (pd.to_datetime('today') - df["BIRTHDATE"]).dt.days // 365
min_age = df['AGE'].min().item()
max_age = df['AGE'].max().item()
print(f"The  age_range is ({min_age}, {max_age})")
# age_range = (, df['AGE'].max())
# print(f"Age range: {age_range.item}")
```

```python
# Number of patients per city/town
pd.set_option('display.max_rows', None)
city_counts = df['CITY'].value_counts()
print(city_counts)
```

```python
# City with highest population
highest_population_city = city_counts.idxmax()
highest_population_count = city_counts.max()
print(highest_population_city)
print(highest_population_count)
```

```python
# City with lowest population
lowest_population_city = city_counts.idxmin()
lowest_population_count = city_counts[city_counts == city_counts.min()]
# Check for zero count cities
zero_count_cities = df['CITY'].isnull().sum()
print(zero_count_cities)

print(lowest_population_count)
```
